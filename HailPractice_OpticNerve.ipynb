**Filters**
* VQSR(Variant Quality Score Recalibration):
    - calculates a new quality score called the VQSLOD (for variant quality score log-odds)
    - uses machine learning algorithms to learn from each dataset what is the annotation profile of good variants vs. bad variants, and does so in a way that integrates information from multiple dimensions. (improvement from original system's binary approach of; is it above or below threshold?)
    - compares variants with 'truth sites' that are provided (ex. HapMap3) to process probability that a variant call is a true genetic variant
    - VQSRTrancheSNP0.00to99: variant was in the range of VQSLODs corresponding to the remaining 0.1% of the truth set, which are considered false positives

**Info**
* BaseQRankSum: REF versus ALT base quality scores. ideal score is close to 0.
* QD:  variant confidence (from the QUAL field) divided by the unfiltered depth of non-hom-ref samples. It is intended to normalize the variant quality in order to avoid inflation caused when there is deep coverage. 
* MQ: mapping quality. when it is good, it is near 60.
* MQRankSum: positive value means the mapping qualities of the reads supporting the alternate allele are higher than those supporting the reference allele (*this is a hard filter, so recommended to be lenient)
* SB: strand bias
* END: end position of the variant described in this record
* ExcessHet: Phred-scaled p-value for exact test of excess heterozygosity. The higher the score, the higher the chance that the variant is a technical artifact or that there is consanguinuity among the samples.
* FS: fisher strand. determines whether there is a difference in the number of reads that support the reference allele and alternate allele on each strand
* InbreedingCoeff: Likelihood-based test for the consanguinuity among samples. 
* MLE: maximum likelihood estimation
* ReadPosRankSum: whether there is evidence of bias in the position of alleles within the reads that support them. ideal score is close to 0
* SOR: estimate strand bias. It was created because FS tends to penalize variants that occur at the ends of exons. It will take into account the ratios of reads that cover both alleles. (*high SOR value: high chance of failing VQSR filter')
* culprit: the annotation that lowers the VQSLOD score the most

**VEP(Variant Effect Predictor)**
* end: end position of the variant described
* nearest: shows gencode transcript of nearest gene
* start: start position of the variant described

**Entry Fields**

* AD: unfiltered allele depth

* DP: filtered depth

* GQ: genotype quality. Phred-scaled confidence that the genotype assignment (GT) is correct. 10log10 p (p: chance that genotype call is wrong)

* GT: genotype. 0: ref, 1: alt, /: unphased, |: phased

* PGT: Physical phasing haplotype information, describing how the alternate alleles are phased in relation to one another. (Which parent is it inherited from?)

* PID: Physical phasing ID information, where each unique ID within a given sample

* PL: "Normalized" Phred-scaled likelihood of possible genotype. Three numbers corresponding to the genotypes (0/0, 0/1, and 1/1). Higher the probability, lower the score.

* PS:  phase set. a set of phased genotypes to which this genotype belongs

* RGQ: Unconditional reference genotype confidence, encoded as a phred quality -10*log10 p(genotype call is wrong)

* SB: strand bias

## Annotating .fam data
We are annotating the .fam data for each sample in order to check each sample's phenotype (is it case or control?) more easily.

```python
faminfo = hl.import_table('CODA_0322.txt', impute=True).key_by('IID')
```

```python
mt = mt.annotate_cols(fam_data = faminfo[mt.s])
```

```python
mt.cols().show()
```

* IID_P: within family ID of father. (0 when father is not in dataset)
* IID_M: within family ID of mother. (0 when mother is not in dataset)
* Sex: 1 = male/ 2 = female/ 0 = unknown
* PHeno_val: 1 = control/ 2 = case/ -9 = missing data

Let's Check how many patients(case) is in our data!

```python
mt.aggregate_cols(hl.agg.counter(mt.fam_data.Pheno_val == 2))
```
